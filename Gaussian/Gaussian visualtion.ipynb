```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_normal(mean=[0],std=[1]):
    for i in range(len(mean)):
        data = np.random.normal(mean[i],std[i],10000)
        sns.kdeplot(data,label = f'm - {mean[i]} | s - {std[i]}')
    plt.legend()
    plt.show()
```

```python
plot_normal([0,1,2,3],[1,.5,2,.2])
```
