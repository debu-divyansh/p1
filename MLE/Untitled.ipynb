```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
import math
```

```python
# We create a population with exponential distribution
population_rate = 3
sample_size = 10000

get_sample = lambda n : np.random.exponential(population_rate,n)
xs = np.arange(0,20,.001)
ys = expon.pdf(xs,scale = population_rate)
```

```python
plt.plot(xs,ys,label ='population')
sample = get_sample(1000)
plt.hist(sample,label = 'sample',density = True)
plt.legend()
plt.show()
```

```python
print(sample.mean())
```

```python
# so we will use MLE to estimate the population parameter using a sampple of the population
log_likelihood = lambda rate : sum(np.log((1/rate)*(math.e**(-v/rate))) for v in sample)
```

```python
rates= np.arange(1,8,0.01)
estimates = [log_likelihood(r) for r in rates]
```

```python
plt.xlabel('parameters')
plt.plot(rates,estimates)
plt.show()
print(rates[estimates.index(max(estimates))])
```
